```python
import random
#The characteristics of the player
class Player:
    def __init__(self):
        self.damage = 10
        self.health = 60
        self.points = 0

    def increase_damage(self, points):
        self.damage += points
        print(f"Attack increased to {self.damage}!")

    def add_points(self, points):
        self.points += points
        print(f"Points earned: {points}, Total points: {self.points}")

#The things that the shop offer to sell during the game
class ShopOwner:
    def __init__(self):
        self.heal_price = 15
        self.poison_price = 20
        self.attack_price = 25
        self.poison_power_increase = 15

    def sell_item(self, player):
        print("Welcome to the shop! Here you can buy items to aid your quest.")
        print("You have", player.points, "points.")

        print("Shop Items:")
        print("1. Heal (Restore health) - 15 points")
        print("2. Poison (Increase attack damage) - 20 points")
        print("3. Increase Attack (Boost attack power) - 25 points")

        choice = input("Enter the number of the item you want to buy: ")
        if choice == '1':
            self.sell_healer(player)
        elif choice == '2':
            self.sell_poison(player)
        elif choice == '3':
            self.sell_damage(player)
        else:
            print("Invalid choice. Please select again.")

    def sell_healer(self, player):
        if player.points >= self.heal_price:
            player.points -= self.heal_price
            player.health += 20
            print(f"You bought a heal! Your health increased to {player.health}.")
            print(f"You now have {player.points} points left.")
        else:
            print("Sorry, you don't have enough points to buy a heal.")

    def sell_poison(self, player):
        if player.points >= self.poison_price:
            player.points -= self.poison_price
            player.increase_damage(self.poison_power_increase)
            print(f"You bought poison! Your attack increased by {self.poison_power_increase}.")
            print(f"You now have {player.points} points left.")
        else:
            print("Sorry, you don't have enough points to buy poison.")

    def sell_damage(self, player):
        if player.points >= self.damage_price:
            player.points -= self.damage_price
            player.increase_damage(20)
            print(f"You increased your attack! Your attack increased.")
            print(f"You now have {player.points} points left.")
        else:
            print("Sorry, you don't have enough points to increase your attack.")

#These are the questions that will randomly assinged in the game
class MathGame:
    def __init__(self):
        self.player = Player()
        self.current_level = 1
        self.game_over = False

    def generate_question(self):
        num1 = random.randint(1, 20)
        num2 = random.randint(1, 20)
        operator = random.choice(['+', '-', '*', '/'])
        if operator == '/':
            answer = "{:.2f}".format(num1 / num2)
        else:
            answer = eval(f"{num1}{operator}{num2}")

        question = f"What is {num1} {operator} {num2}? "
        return question, str(answer)

    def validate_answer(self, user_answer, correct_answer):
        return user_answer == correct_answer

    def damage_enemy(self, enemy):
        enemy_health = enemy['health']
        enemy_damage = enemy['damage']
        while enemy_health > 0 and not self.game_over:
            question, correct_answer = self.generate_question()
            print(question)
            user_answer = input("Your answer: ")

            if self.validate_answer(user_answer, correct_answer):
                enemy_health -= self.player.damage
                self.player.health -= enemy_damage
                print(f"You attacked the {enemy['name']} for {self.player.damage} damage!")
                print(f"{enemy['name']} attacks back for {enemy_damage} damage!")
                print(f"Player Health: {self.player.health}, {enemy['name']} Health: {enemy_health}")
                if self.player.health <= 0:
                    print("You've been defeated!")
                    self.game_over = True
                elif self.current_level == 3 and enemy_health <= 0:
                    print("Congratulations! You defeated the Dragon! You saved the kingdom!")
                    self.game_over = True

            elif enemy_health <= 0:
                print(f"You defeated the {enemy['name']}!")
                self.player.add_points(enemy['reward'])

    def visit_shop(self, player, shop):
        print("Welcome to the shop! Here you can buy items to aid your quest.")
        print("You have", player.points, "points.")

        choice = input("Would you like to buy poison to increase attack? (y/n): ")
        if choice.lower() == 'y':
            shop.sell_poison(player)
        else:
            print("Thanks for visiting the shop!")

# This explain the game to the player
    def play_game(self):
        print("You find yourself on a quest to save the kingdom!")
        print("You are our hero and Math is your best subject.")
        print("You can choose to defeat the enemies to earn points.")
        print("You can kill as many goblins as you want before moving to the next level.")
        print("Please don't forget to look at your health before moving to your next step.")
        print("Let's start now, my hero.")

        while not self.game_over:
            print(f"\nLevel {self.current_level}")
            choice = input("Choose an option: \n1. Start this level\n2. Visit shop\n3. Next level\nEnter option number: ")
            if choice == '1':
                if self.current_level == 1:
                    goblin = {'name': 'Goblin', 'health': 5, 'damage': 5, 'reward': 10}
                    self.damage_enemy(goblin)
                    if self.player.health > 0:
                        self.player.add_points(10)
                elif self.current_level == 2:
                    wizard = {'name': 'Wizard', 'health': 7, 'damage': 20, 'reward': 20}
                    self.damage_enemy(wizard)
                    if self.player.health > 0:
                        self.player.add_points(20)
                elif self.current_level == 3:
                    dragon = {'name': 'Dragon', 'health': 12, 'damage': 50, 'reward': 500}
                    self.damage_enemy(dragon)
                    if self.player.health > 0:
                        self.player.add_points(50)

            elif choice == '2':
                shop = ShopOwner()
                shop.sell_item(self.player)
            else:
                self.current_level += 1

# Start the game
game = MathGame()
game.play_game()
```
